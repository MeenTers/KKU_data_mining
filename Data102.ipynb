##BarChart

```
import matplotlib
import matplotlib.pyplot as plt  #เรียกใช้การวาดกราฟ
import numpy as np
```

```
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]          #กำหนดตัวแปร
women_means = [25, 32, 34, 20, 25]
```

```
x = np.arange(len(labels))  # the label locations  
width = 0.35  # the width of the bars
```

```
plt.bar(x - width/2, men_means, width, label='Men')      #คำสั่ง.bar สามารถสร้างกราฟแท่งได้แต่ไม่สามารถเปลี่ยนแปลงค่าแกนxได้
plt.bar(x + width/2, women_means, width, label='Women')
plt.ylabel('scores')
plt.title('Scores by group and gender')
plt.legend();
```

```
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')            #คำสั่งวาดboxplot มาจากgoogle
rects2 = ax.bar(x + width/2, women_means, width, label='Women')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
#สองบรรทัดนี้มาด้วยกัน
ax.set_xticks(x)
ax.set_xticklabels(labels);
ax.legend()
```
